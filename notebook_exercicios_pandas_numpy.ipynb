## 8. Criando novas colunas

### Exercício 16
Crie uma coluna chamada `media` com a média entre `nota1` e `nota2`.

### Exercício 17
Crie uma coluna chamada `aprovado`:
- `True` para média maior ou igual a **7**
- `False` caso contrário

### Exercício 18
Crie uma coluna chamada `situacao_faltas`:
- `Alto risco` para mais de 10 faltas
- `Atenção` para 6 a 10 faltas
- `Ok` para até 5 faltas


```python
df['media'] = (df['nota1'] + df['nota2']) / 2
df[['nome', 'nota1', 'nota2', 'media']].head()
```

```python
df['aprovado'] = df['media'] >= 7
df[['nome', 'media', 'aprovado']].head()
```

```python
import numpy as np
condicoes = [
    (df['faltas'] > 10),
    (df['faltas'] >= 6) & (df['faltas'] <= 10),
    (df['faltas'] <= 5)
]
escolhas = ['Alto risco', 'Atenção', 'Ok']
df['situacao_faltas'] = np.select(condicoes, escolhas, default='N/A')
df[['nome', 'faltas', 'situacao_faltas']].head(10)
```

```python
emails_invalidos = df_limpo[
    (~df_limpo['email'].str.contains('@', na=False)) |
    (~df_limpo['email'].str.contains('.com', na=False))
]

print("Registros com e-mails possivelmente inválidos (incluindo os vazios):")
emails_invalidos[['nome', 'email']]

```

## 12. Introdução ao NumPy

### Exercício 27
Crie um array NumPy com os valores da coluna `nota1`, removendo os nulos.

### Exercício 28
Com esse array, calcule:
- média
- maior valor
- menor valor
- soma

### Exercício 29
Some **1 ponto** em todas as notas do array.

### Exercício 30
Mostre apenas as notas maiores que **7** usando NumPy.


```python


import numpy as np
notas_array = df_limpo['nota1'].dropna().to_numpy()
print("Array de notas criado com sucesso!")
print(notas_array)

```

```python
media = np.mean(notas_array)
maior = np.max(notas_array)
menor = np.min(notas_array)
soma = np.sum(notas_array)
print(f"Média: {media:.2f}")
print(f"Maior nota: {maior}")
print(f"Menor nota: {menor}")
print(f"Soma total: {soma}")
```

```python
notas_bonus = notas_array + 1
print("Notas com bônus de 1 ponto:")
print(notas_bonus)
```

```python
notas_altas_np = notas_array[notas_array > 7]

print("Notas maiores que 7 (via NumPy):")
print(notas_altas_np)
```
